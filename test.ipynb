```python
ticket_number = 123456

ticket_str = str(ticket_number)
sum_first_part = sum(int(number) for number in ticket_str[:3])
sum_last_part = sum(int(number) for number in ticket_str[3:])

print("Счастливый билет!") if sum_first_part == sum_last_part else print("Несчастливый билет")
```
